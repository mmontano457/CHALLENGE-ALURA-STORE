### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda
tienda2
tienda3
tienda4

tienda.head()
```

#1. Análisis de facturación



```python
facturacion1=tienda['Precio'].sum()
facturacion2=tienda2['Precio'].sum()
facturacion3=tienda3['Precio'].sum()
facturacion4=tienda4['Precio'].sum()
print(facturacion1)
print(facturacion2)
print(facturacion3)
print(facturacion4)
```

```python
total_fact=[]
total_fact.append(facturacion1)
total_fact.append(facturacion2)
total_fact.append(facturacion3)
total_fact.append(facturacion4)
print(total_fact)

tiendas=[1,2,3,4]

plt.bar (x=tiendas, height=total_fact)
plt.show()
```

Graficamente se destaca que la tienda que mas facturó en el periodo enero 2020- marzo 2023 fue la tienda 1. La que menos vendio la tienda 4.

El gráfico nos muestra que la tienda mejor calificada es la 3 y la tienda 1 la que tiene menor calificación.
La escala de calificación es de 1 a 5, y el promedio de las tiendas todas rondas el 4, por lo que no se puede decir que existe una gran diferencia entre ellas en las calificaciones.

# 5. Envío promedio por tienda

```python
envio_prom1=tienda ['Costo de envío'].mean()
envio_prom2=tienda2['Costo de envío'].mean()
envio_prom3=tienda3['Costo de envío'].mean()
envio_prom4=tienda4['Costo de envío'].mean()
print(envio_prom1)
print(envio_prom2)
print(envio_prom3)
print(envio_prom4)
```

```python
envio_prom_tienda=[]
envio_prom_tienda.append(envio_prom1)
envio_prom_tienda.append(envio_prom2)
envio_prom_tienda.append(envio_prom3)
envio_prom_tienda.append(envio_prom4)
envio_prom_tienda


tiendas=[1,2,3,4]
x=tiendas
y=envio_prom_tienda

plt.bar(x, y)
plt.show()
```

El promedio de costo de envío de las tiendas, ronda los 2500, siendo la que tiene un mayor costo de envio promedio la tienda 1, y la con menor costo de envio promedio la tienda 4.

```python
tienda['Ganancia']=(tienda['Precio']-tienda['Costo de envío'])
tienda2['Ganancia']=(tienda['Precio']-tienda['Costo de envío'])
tienda3['Ganancia']=(tienda['Precio']-tienda['Costo de envío'])
tienda4['Ganancia']=(tienda['Precio']-tienda['Costo de envío'])

Utilidad1=tienda['Ganancia'].sum()
Utilidad2=tienda2['Ganancia'].sum()
Utilidad3=tienda3['Ganancia'].sum()
Utilidad4=tienda4['Ganancia'].sum()

Utilidad_por_tienda=[]
Utilidad_por_tienda.append(Utilidad1)
Utilidad_por_tienda.append(Utilidad2)
Utilidad_por_tienda.append(Utilidad3)
Utilidad_por_tienda.append(Utilidad4)

Utilidad_por_tienda
```

Considerando solo la informacion que tenemos (por lo que estamos obviando otros aspectos de costos directo), y generando una nueva variable llamada Utilidad, que es el resultado de restar a lo facturado (asumiendo que Precio es lo facturado) el Costo de Envío, nos encontramos con que tres de las 4 tiendas tienen la misma utilidad de $1089502500 y la tienda 4 una utilidad de $1089316100, siendo esta menor. Considerando solo esta información, se recomienda vender las tienda 4.
